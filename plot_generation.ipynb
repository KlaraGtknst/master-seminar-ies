```python
import matplotlib.pyplot as plt
import numpy as np

image_src = 'images/'
"""Plot exponential functions."""
x = np.linspace(-1, 2, 100)

# negative loss
y = np.exp(-x) + min(-np.exp(-x))   # TODO: comment second term
plt.figure()
plt.plot(x, y, label='Negative', color='orange')
plt.xlabel("Distance of anchor and generated sample")
plt.ylabel("Loss")

# positive loss
plt.plot(x, -np.exp(-x), label='Positive', color='green')   # TODO: + 1

# small distance
plt.scatter(-0.75, np.exp(0.75) + min(-np.exp(-x)), label=r'Hard negative $x^-$', color='red')
plt.scatter(-0.75, -np.exp(0.75), label=r'Easy positive $x^+$', color="cyan")

# large distance
plt.scatter(1.75, np.exp(-1.75) + min(-np.exp(-x)), label=r'Easy negative $x^-$', color='pink')
plt.scatter(1.75, -np.exp(-1.75), label=r'Hard positive $x^+$', color="violet")

plt.tick_params(
    axis='both',          # changes apply to the x-axis
    which='both',      # both major and minor ticks are affected
    bottom=False,      # ticks along the bottom edge are off
    top=False,         # ticks along the top edge are off
    left=False,
    right=False,
    labelleft=False,
    labelbottom=False) # labels along the bottom edge are off
plt.legend()
plt.savefig(image_src + 'Hard_easy_samples_dist_effect_loss.svg')
plt.show()
```
